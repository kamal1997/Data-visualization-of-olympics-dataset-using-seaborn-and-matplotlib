```
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```
df_ath = pd.read_csv('athlete_events.csv')
```

```
df_ath.head(10)
```

```
df_ath['total'] = df_ath['Gold_Medal']+df_ath['Silver_Medal']+df_ath['Bronze_Medal']
```

```
df_ath.head(10)
```

```
df_ath.shape
```

```

df_main = df_ath[['Team','total']]
df_main.head()
```

```
 df_group = df_main.groupby('Team',axis=0).sum()
 df_group = df_group.sort_values(by='total', ascending=False)
 df_top10 = df_group.head(20)
```

```
import matplotlib.pyplot as plt
df_top10.plot(kind="bar")
plt.title("No of medals of Top 20 countries from 1896 to 2016")
plt.xlabel("Countries")
plt.ylabel("No.of medals")
```

```
df_r = df_ath[['Team','total','Year']]
df_in = df_r.loc[df_r['Team'] == "India"]
```

```
df_in = df_in.groupby('Year').sum()
df_in.plot(kind='line')
plt.title("No.of medals India got in Olympics from 1896 to 2016")
plt.xlabel("Years")
plt.ylabel("No of medals")
```

```
df_india = df_ath[['Team','Sport','total']]
df_india = df_india.groupby('Sport').sum()
df_india = df_india.sort_values(by='total', ascending=False)
df_india1 = df_india.head(10)
df_india = df_india.head(5)
df_india['total'].plot(kind='pie')
plt.title('Top sports in India that earned medals in olympics since 1896')

```

```
df_1990 = df_ath[df_ath['Year']>1989]

df_1990i = df_1990.sort_values(by='Year', ascending=False)
df_1990i = df_1990i[df_1990i['Team']=='India']
df_1990i = df_1990i[['Year','total']].groupby('Year').sum()
df_1990i['total']

df_all = df_1990i
df_all = df_all.rename(columns={'total':'India'})


df_1990c = df_1990.sort_values(by='Year', ascending=False)
df_1990c = df_1990c[df_1990c['Team']=='China']
df_1990c = df_1990c[['Year','total']].groupby('Year').sum()

df_1990c.rename(columns={'total':'China'})

df_all['China'] = df_1990c['total']

df_1990u = df_1990.sort_values(by='Year', ascending=False)
df_1990u = df_1990u[df_1990u['Team']=='United States']
df_1990u = df_1990u[['Year','total']].groupby('Year').sum()

df_1990c.rename(columns={'total':'United States'})

df_all['United States'] = df_1990u['total']
df_all

df_all.plot()
plt.xlabel('Year')
plt.ylabel('No of medals')
plt.title('No of medals of India vs US vs China')

```
